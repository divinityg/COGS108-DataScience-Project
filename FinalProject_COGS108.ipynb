# Research Question

The avocado, or persea americana, is the emblematic fruit of California, and certainly amongst us residents in San Diego county (one of two major avocado-producing counties in the state). California is not without its droughts, however, which makes us climate-conscious individuals—with a love for the dark-skinned, bright green fruit—curious as to whether the lack or excess of rainfall may have any impact on the production of avocados in San Diego. Can precipitation level be used to model average San Diego avocado market price in a given year, using data from 1996-2022?

# Hypothesis


Our null hypothesis is that precipitation has no significant impact on the quantity of avocados produced leading to no price change. However, we believe that during avocado season, there exists an interval of precipitation values in which avocado production is heightened and drives down the price. If rainfall is too high, crops are oversaturated with water and will not grow — if rainfall is too low, crops are undersaturated with water and will not grow.

# Data

To investigate the relationship between precipitation and avocado prices in San Diego, an ideal dataset would include measurements of precipitation (inches) per month<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1), the average market price of avocados per month, and the amount of avocado production<a name="cite_ref-1"></a>[<sup>2</sup>](#cite_note-2). Precipitation would be the independent variable, and avocado market price would be the dependent variable in this study. A potential confound that may exist in the dataset would be farmers watering their crops independent of naturally occurring rainfall. Another consideration is the demand of avocados which can also impact price. We would be selecting years before and after the avocado toast trend to acquire more data, and we assumes demand to be unchanged over the years 1996-2022<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Additionally, temperature, wildfire, disease, and policy changes could also affect avocado production in a given period. Ideally we would collect data from different months over the course of several years to ensure a large enough sample size of rainfall and avocado prices to test our hypothesis.

We have found datasets of monthly rainfall levels and temperature by county in California, provided by National Centers for Environmental Information<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). ## San Diego County Precipitation 1996-2022

```python
import pandas as pd
import numpy as np
```

```python
# read all datasets from year 1996-2022 and store in a list
pcps = list()
for year in range(1996, 2023):
    # remove the overhead rows by setting header = 4
    pcps.append(pd.read_csv(f'https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/county/mapping/4-pcp-{year}01-12.csv', header=4))
```

```python
# concatenate the rows of the dataframes that uses the years as key
pcp = pd.concat(pcps, keys=range(1996, 2023))
# convert the keys to columns using reset_index without dropping
pcp = pcp.reset_index(drop=False)
pcp
```

```python
# rename the level_0 column to Year
pcp = pcp.rename(columns={'level_0':'Year'})
pcp
```

```python
# drop unnecessary columns
cleaned_pcp = pcp.drop(['level_1', 'ID', 'State', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean'], axis=1)
cleaned_pcp
```

```python
# filter for only San Diego data
cleaned_pcp = cleaned_pcp.query('Name == "San Diego County"').reset_index(drop=True)
cleaned_pcp
```

```python
# remove Name column, rename Value to Precipitation
cleaned_pcp = cleaned_pcp.drop(['Name'], axis=1).rename({'Value':'Precipitation'}, axis=1)
cleaned_pcp
```

```python
# check data types
cleaned_pcp.dtypes
```

## San Diego County Temperature 1996-2022

```python
# drop unnecessary columns
cleaned_temp = temp.drop(['level_1', 'ID', 'State', 'Rank', 'Anomaly (1901-2000 base period)', '1901-2000 Mean'], axis=1)
cleaned_temp
```

## San Diego County Crop Report 1996-2022

```python
avocado = pd.read_csv('Dataset\San Diego County Crop Report 1996-2022.csv')
avocado
```

```python
# check data type
avocado.dtypes
```

```python
# define function to remove ',' and '$' and convert to integer
def stringtoint(string):
    out = string.replace(',', '')
    out = out.replace('$', '')
    return int(out)
```

```python
# change columns to integer
avocado['Acres (Harvested)'] = avocado['Acres (Harvested)'].apply(stringtoint)
avocado['Tons (Produced)'] = avocado['Tons (Produced)'].apply(stringtoint)
avocado['Total Amount ($)'] = avocado['Total Amount ($)'].apply(stringtoint)
avocado
```

```python
# calculate $/Ton column
avocado['$/Ton'] = avocado['Total Amount ($)'].div(avocado['Tons (Produced)'])
avocado
```

```python
# round $/Ton to 2 decimal places
avocado['$/Ton'] = avocado['$/Ton'].round(2)
avocado
```

### Regressional Analysis

We will attempt to determine the strength of the relationship between avocado price/produced based on independent variable such as precipitation and temperature in fahrenheit.

First, let's check if there's a linear relationship between `Tons (Produced)` and `Total Amount ($)`.

```python
sns.regplot(df, x='Total Amount ($)', y='Tons (Produced)')
plt.title('Linear Regression');
```

```python
import patsy
import statsmodels.api as sm
```

```python
outcome, predictors = patsy.dmatrices("Q('Tons (Produced)') ~ Q('Total Amount ($)')", df)
res_1 = sm.OLS(outcome, predictors).fit()
print(res_1.summary())
```

Even though the `R-squared` value is relatively low, there is definitely a linear relationship between `Tons (Produced)` and `Total Amount ($)` as the p-value is 0.0005 < 0.01.

Going forward, we will just predict `Total Amount ($)`, since it is part our hypothesis and predicting `Total Amount ($)` would give us information on `Tons (Produced)`.

Let's start with a linear regression between `Total Amount ($)` and `Precipitation`.

```python
sns.regplot(df, x='Precipitation', y='Total Amount ($)')
plt.title('Linear Regression');
```

The shaded region of the graph represents a 95% confidence interval that the actual slope of the regression line lies within the shaded part of the plot.

The confidence interval includes the horizontal line, meaning that there might not be linear relationship between `Precipitaion` and `Total Amount ($)`.

Let's look at a more detailed regression result using statsmodels.

```python
outcome, predictors = patsy.dmatrices("Q('Total Amount ($)') ~ Precipitation", df)
res_1 = sm.OLS(outcome, predictors).fit()
print(res_1.summary())
```

From the regression summary table, we can see that the `R-squared` and `Abj. R-squared` values of this model is very low which means that this model is not a good fit of the data.

The `Prob (F-statistic)` is 0.0837 which is greater than 0.01 indicating that there is a decent likelihood of a linear relationship arising due to chance. 

Whereas a less than 0.01 `Prob (F-statistic)` means the likelihood of a linear relationship is significant.

This result is what we expected, as we hypothesized that there exists an interval of precipitation values in which avocado production is heightened and drives down the price and outside the interval avocado production decreses. 

So we will experiment with non-linear models, such as second order regression, to represent the relationship.

Another idea is to split up the data into two, below the mean precipitation and above the mean precipitation. If our hypothesis is correct, then there could be a positive linear relationship below the mean precipitaion and a negative linear relationship above the mean precipitation.

### Decision Tree

Decision tree is a model where the data is split into branches at certain thresholds, given a max depth as the hyperparameter. 

Let's try using decision tree and check for their performance. 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import LeaveOneOut
```

```python
# set up the variables, X is precipitation, X2 is temperature, y is total amount
X = df['Precipitation'].to_numpy().reshape(-1, 1)
X2 = df['Fahrenheit'].to_numpy().reshape(-1, 1)
y = df['Total Amount ($)'].to_numpy().reshape(-1, 1)
# split up data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X2_train, X2_test = train_test_split(X2, test_size=0.2, random_state=0)
```

#### Max Depth 2

```python
DTregressor = DecisionTreeRegressor(max_depth=2)
DTregressor.fit(X_train, y_train)
print(f'Training set score: {DTregressor.score(X_train, y_train)}')
print(f'Test set score: {DTregressor.score(X_test, y_test)}')
```

The score here is the same R-squared as above.
The training set is expected to have a high score.
The test set is negative as this model is worse at predicting then just taking the mean of y values.

Let's plot the data to see the result.

```python
y_train_pred = DTregressor.predict(X_train)
y_test_pred = DTregressor.predict(X_test)
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.scatter(X_train, y_train_pred, color="cornflowerblue", label="train")
plt.scatter(X_test, y_test_pred, color="darkblue", label="test")
plt.xlabel("Precipitation")
plt.ylabel("$")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

Maybe max depth of 2 is not enough to represent the trend of the data.
Let's try with bigger depth.

#### Max Depth 3

```python
DTregressor = DecisionTreeRegressor(max_depth=3)
DTregressor.fit(X_train, y_train)
print(f'Training set score: {DTregressor.score(X_train, y_train)}')
print(f'Test set score: {DTregressor.score(X_test, y_test)}')
```

```python
y_train_pred = DTregressor.predict(X_train)
y_test_pred = DTregressor.predict(X_test)
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.scatter(X_train, y_train_pred, color="cornflowerblue", label="train")
plt.scatter(X_test, y_test_pred, color="darkblue", label="test")
plt.xlabel("Precipitation")
plt.ylabel("$")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
